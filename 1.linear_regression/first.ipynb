# Absolute Trick
$$
q = w_1*x + w_2
$$
add p to the slop and 1 the intercept
$$
q =(w_1 + p)*x + (w_2 +1)
$$
add learning rate alpha Î±
$$
y = (w_1 +p \alpha)x + (w_2 + \alpha)
$$
when the point is underneath the line we need to 
subtract to move the line appropriately 
$$
y = (w_1 - p \alpha)x + (w_2 - \alpha)
$$


```python
import matplotlib.pyplot as plt
import numpy as np


x_dot = 5  
y_dot = 10   
plt.scatter(x_dot, y_dot, color='red', label=f'Point ({x_dot}, {y_dot})', s=50)  


slope = 0.5  
intercept = 0  
x = np.linspace(-10, 10, 1000)  

y = slope * x + intercept

plt.plot(x, y, label=f'y = {slope}x + {intercept}', color='black')

p = x_dot
slope2 = slope + p
intercept2 = intercept + 1
y = slope2 * x + intercept2

plt.plot(x, y, label=f'y2 = {slope2}x + {intercept2}', color='red')
# we see we make big setp that not what we want so let's make smaller steps
# so let's add Î± range rate variable let's it's equal now 0.1
Î± = 0.1
slope3 = slope + p * Î±
intercept3 = intercept + Î±
y = slope3 * x + intercept

plt.plot(x, y, label=f'step1 = {slope3}x + {intercept3}', color='yellow')

slope3 += (p * Î±)
intercept3 += Î±
y = slope3 * x + intercept
plt.plot(x, y, label=f'step2 = {slope3}x + {intercept3}', color='Chartreuse')
# and another step
slope3 += (p * Î±)
intercept3 += Î±
y = slope3 * x + intercept
plt.plot(x, y, label=f'step3 = {slope3}x + {intercept3}', color='green')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a Line with Slope')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

# Square Trick
if you notic that the previouse method don't help here <br>
becuase it's depeand on the horizantal destince <br>

<img src="squarTrick1.png" width=300>
<hr>
<br>
so we will use another method called **square trick**

$$
y = \left( w_1 + p(q - q') \alpha \right) x + \left( w_2 + (q - q') \alpha \right)
$$

<img src="squarTrick2.png" width=700>

```python
x_dot = -5
y_dot = 3

plt.scatter(x_dot, y_dot, label=f'Point ({x_dot}, {y_dot})', color='red')

x = np.linspace(-10, 10, 1000)  
slope = -0.6
intercept = 4
y = slope * x + intercept
p = x_dot
q = y_dot
alpha = 0.01
plt.plot(x, y, label=f'base line = {slope}x + {intercept}', color='black')

counter = 1

# Update loop to recalculate Q after slope and intercept adjustment
while True:
    Q = slope * x_dot + intercept
    print(Q)
    if counter >= 20:
        break  # Exit loop if the line has moved to or above the target point
    slope += p * (q - Q) * alpha
    intercept += (q - Q) * alpha
    y = slope * x + intercept
    plt.plot(x, y, color=blue_colors[counter % len(blue_colors)])
    counter += 1

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a Line with Slope')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

# Mean Absolute Error


```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
tests = pd.read_csv('Linear RegressionSheet1.csv')
pred_ys = []
def fn (x):
    fn1 =  2 * x + 1
    return fn1
mean = 0
for i in range(300):
    x = int(tests['X'][i])
    pred_y = fn(x)
    pred_ys.append(pred_y)
    y = int(tests['Y'][i])
    dif = abs(y - pred_y)
    mean += dif
mean /= 300
mean2 = mean_absolute_error(tests['Y'],pred_ys)
print(mean2)
print(mean)    
```
